```python
# doubly ended LL
from collections import deque

class DoublyEndedLinkedListNode:
    def __init__(self, key, val, prev=None, nxt=None):
        self.prev = nxt
        self.next = prev
        self.key = key
        self.val = val

    def __str__(self):
        return str(self.val)

class LRUCache:
    def __init__(self, capacity: int):
        self.capacity = capacity or 1
        self.hash_data = {}
        self.head = None
        self.tail = None

    def remove_tail(self):
        if not self.tail:
            return

        if self.tail == self.head:
            self.tail = self.head = None
            return

        old_tail = self.tail
        new_tail = old_tail.prev
        new_tail.next = None
        self.tail = new_tail

    def evict(self):
        item_to_delete = self.tail
        self.remove_tail()

        del self.hash_data[item_to_delete.key]

    def update_node_to_head(self, node_to_process):
        node = node_to_process

        # already at the head
        if node == self.head:
            return

        # case where have no items in our DLL
        if not self.head or not self.tail:
            self.head = self.tail = node
            return


        if self.tail == node:
            self.remove_tail()

        node.next = None
        node.prev = None

        self.head.prev = node
        node.next = self.head

        # finally set head to this newly accessed element
        self.head = node

    def get(self, key: int) -> int:
        if key not in self.hash_data:
            return -1

        node_to_process = self.hash_data[key]
        self.update_node_to_head(node_to_process)

        return node_to_process.val

    def put(self, key: int, value: int) -> None:
        if key in self.hash_data:
            node_to_process = self.hash_data[key]
            node_to_process.val = value
        else:
            node_to_process = DoublyEndedLinkedListNode(key=key, val=value)
            self.hash_data[key] = node_to_process

        # delete last element 
        if len(self.hash_data) > self.capacity:
            self.evict()

        self.update_node_to_head(node_to_process)
```
