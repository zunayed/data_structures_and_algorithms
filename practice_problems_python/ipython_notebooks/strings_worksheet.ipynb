## Is Palindrome

```python
"""
start 4:44
end 4:54


"A man, a plan, a canal: Panama"
  p1                            
                              p2

Time - O(n)
space - O(1)

"""

class Solution:
    def isPalindrome(self, s: str) -> bool:
        p1, p2 = 0, len(s) - 1

        while p1 < p2:
            # print(s[p1], s[p2])
            if not s[p1].isalnum():
                p1 += 1
                continue
        
            if not s[p2].isalnum():
                p2 -= 1
                continue
            

            if s[p1].lower() != s[p2].lower():
                return False
            
            p1 += 1
            p2 -= 1
        
        return True
```

## Longest substring without repeating characters
https://leetcode.com/problems/longest-substring-without-repeating-characters/

```python
"""
start 1:00
end 1:25

Input: "abcabcbb"
          s i
        [bzykr]

Input: "abcabcbbzykr"
"""

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        if not s:
            return 0 
        
        longest_seen = 1
        start = -1
        
        letter_set = set() 
        
        for i in range(len(s)):
            
            while s[i] in letter_set:
                start += 1
                letter_set.remove(s[start])
            
            letter_set.add(s[i])
            longest_seen = max(longest_seen, (i - start))
        
        return longest_seen 
```
