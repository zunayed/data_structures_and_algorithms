## Quicksort

start on: Sunday 12/08/2019 - **2:15pm**   
end on:   Sunday 12/08/2019 - **2:35**

Time ~= `O(nlogn)`  
Space = `O(1)`

````python
def quickSort(array):
    def partition(begin, end):
        pivot = begin

        for i in range(begin + 1, end + 1):
            if array[i] < array[begin]:
                pivot += 1
                
        array[i], array[pivot] = array[pivot], array[i]
        array[pivot], array[begin] = array[begin], array[pivot]
        return pivot

def _qs(begin, end):
    if begin >= end:
        return

    pivot = partition(begin, end)
    _qs(begin, pivot - 1)
    _qs(pivot + 1, end)

    _qs(0, len(array) - 1)

    return array
````

````python
class Solution:
    def insertionSortList(self, head):
        if not head:
            return 
        
        p1 = head
        p2 = head.next
        
        smallest = head

        while p1:
            while p2: 
                if p2.val < smallest.val:
                    smallest = p2 
                    
                p2 = p2.next
                
            # swap
            p1.val, smallest.val = smallest.val, p1.val
            p1 = p1.next
            
            smallest = p1
            p2 = p1
            
        return head
````

## Sort-colors
https://leetcode.com/problems/sort-colors/

start on: Sunday 12/08/2019 - **10:00am**   
end on:   Sunday 12/08/2019 - **10:45**

Time ~= `O(n)`  
Space = `O(1)`

````python
RED = 0
WHITE = 1
BLUE = 2

class Solution:
    def sortColors(self, nums):
        p1 = 0                  # end of red colors
        curr = 0                
        p2 = len(nums) - 1      # start of blue colors

        while curr <= p2:
            if nums[curr] == RED:
                # swap curr and p1 and move p1 forward
                nums[curr], nums[p1] = nums[p1], nums[curr]
                p1 += 1
                curr += 1
            elif nums[curr] == BLUE:
                # swap curr and p2 and move p2 back
                nums[curr], nums[p2] = nums[p2], nums[curr]
                p2 -= 1
            else:   
                curr += 1
````
